```python
import pandas as pd
df = pd.read_csv('C:/Users/renan/OneDrive/Documentos/Python Scripts/AquivosAnalise/Marketing Campaigns.csv',sep=';')
df = df.drop(['customer-id'], axis=1)
df = df.head(8)
```

```python

# eixo_y.reset_index(inplace=True)

from matplotlib import  legend, pyplot as plt
y=df.drop(['location'],axis=1)
x=df.location
plt.figure(figsize=None)
plt.plot(x,y,label='dados',ls='--',lw=2,marker='+')
plt.title('Meu grafico')
plt.xlabel('localizacoes')
plt.ylabel('line')
plt.legend()

plt.show()


```

```python
figura = plt.figure(figsize=(10,4))
figura.suptitle('geral')

figura.add_subplot(131)
plt.plot(x,y,ls='--',lw=2,marker='+')
plt.title('grafico 1')

figura.add_subplot(132)
plt.bar(df.age,df.location)
plt.title('grafico 2')

figura.add_subplot(133)
plt.scatter(df['email-opened'],df.location)
plt.title('grafico 3')
```
