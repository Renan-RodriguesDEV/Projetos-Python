```python
import pandas as pd
import plotly.express as px

# Carregando os dados do arquivo CSV
tbl = pd.read_csv('C:/Users/renan/OneDrive/Documentos/Python Scripts/Machine_Learning/exemplo2.csv')

#criando um dataframe
df = pd.DataFrame(tbl)
display(df.head(7))
display(df.describe())
```

```python
# criando um dicionario
dicionario = {
    'nome':['Jose','maria','amaral'],
    'idade':[19,19,45],
    'salario':[1000,20000,3050]
}
for linha in dicionario:
    print(dicionario[linha])
```

```python
# traformando em dataframe
df_dicionario = pd.DataFrame(dicionario)
display(df_dicionario)
```

```python
x_dicionario = df_dicionario['idade']
y_dicionario = df_dicionario['salario']

fig = px.line(df_dicionario,x=x_dicionario,y=y_dicionario,title='Salario por idade',line_shape=None)


fig.show()
```
